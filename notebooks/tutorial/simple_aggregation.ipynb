```python
import torch

# 節點 0,1,2,3
h = torch.tensor([
    [11, 11],   # h[0]
    [21, 21],   # h[1]
    [31, 31],   # h[2]
    [41, 41]    # h[3]
])  # 形狀 [4, 2]

edge_index = torch.tensor([
    [0, 0, 1, 2, 2],   # row (source) → 中心節點
    [1, 2, 0, 0, 3]    # col (target) → 鄰居
])  # 邊：0←1, 0←2, 1←0, 2←0, 2←3

row, col = edge_index
num_neigh = torch.bincount(row, minlength=h.size(0)).float()
print("Row tensor:", row)
print("Number of neighbors for each node:", num_neigh)
```

```python
aggr = torch.zeros_like(h)   # [[0,0], [0,0], [0,0], [0,0]]
print("Before aggregation:\n", aggr)
print(aggr.shape)  # Should be 4, the number of nodes

aggr.index_add_(0, row, h[col])
print("After aggregation:\n", aggr)
```
