```python
import pandas as pd

df = pd.DataFrame({
    'A': [10, 20, 30, 40],
    'B': ['a', 'b', 'c', 'd'],
    'C': [0.1, 0.2, 0.3, 0.4]
}, index=['r0', 'r1', 'r2', 'r3'])

print(df)
```

```python
row_2 = df.iloc[1]
print(row_2)
```

```python

col_0 = df.iloc[:, 0]
print(col_0)

```

```python
val = df.iloc[1, 2]
print(val)  # 0.2
```

```python
sub = df.iloc[0:2, 0:2]
print(sub)
```

```python
sel = df.iloc[[0, 2], [1]]
print(sel)
```

```python
last_row = df.iloc[-1]
print(last_row)
```

```python
mask_rows = [True, False, True, False]
filtered = df.iloc[mask_rows]
print(filtered)
```
