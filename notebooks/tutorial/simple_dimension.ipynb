```python
import torch

A = torch.tensor([
    [10, 20, 30],    # row 0
    [40, 50, 60],    # row 1
    [70, 80, 90]     # row 2
])   # shape (3, 3)

print(A.sum(dim=0))   # → tensor([120, 150, 180])   ← added **down** each column
print(A.sum(dim=1))   # → tensor([ 60, 150, 240])   ← added **across** each row
```
