```python
import torch

print("=== Basic .index_add_() example ===\n")

# ────────────────────────────────────────────────
# Situation: we want to accumulate values into specific positions
# ────────────────────────────────────────────────

# Target tensor (will receive the additions)
target = torch.zeros(8)          # like 8 buckets / classes / positions
print("Initial target:", target)

# Indices telling us **where** to add each value
indices = torch.tensor([1, 4, 1, 7, 4, 1])    # ← position 1 gets many hits

# Values we want to add
values = torch.tensor([10., 20., 30., 5., 15., 40.], dtype=torch.float32)

# ────────────────────────────────────────────────
# Method 1: using index_add_ (in-place)
# ────────────────────────────────────────────────

target.index_add_(dim=0, index=indices, source=values)

print("\nAfter index_add_:")
print(target)
# Expected: position 1 ← 10+30+40 = 80
#           position 4 ← 20+15     = 35
#           position 7 ← 5
```
