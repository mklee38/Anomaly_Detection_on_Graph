```python
import torch
import torch.nn as nn

# Create a linear layer: 20 → 30
linear = nn.Linear(in_features=2, out_features=3)   # in_features 是輸入特徵的維度，out_features 是輸出特徵的維度
print(linear)                #   → Linear(in_features=2, out_features=3, bias=True)

# Typical usage
x = torch.randn(5, 2)          # batch size = 5, feature dimension = 2
print("x is equal to", x)           # → tensor([[-0.1234,  0.5678],   

y = linear(x)           # → tensor([[ 0.2345, -0.6789,  1.2345],
print("(3)", y.shape)                    # → torch.Size([5, 3])

# Single sample
x_single = torch.randn(2)   # feature dimension = 2
y_single = linear(x_single)     # → tensor([ 0.1234, -0.5678,  1.2345])

print("(4)", x_single.shape)  # print shape of single input (expected torch.Size([2]))
print("(5)", y_single.shape)             # print shape of single output (expected torch.Size([3]))
```
